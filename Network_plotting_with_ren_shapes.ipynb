```python
import geopandas as gpd
import geoplot
from shapely.ops import unary_union
import shapely
import matplotlib
import matplotlib.pyplot as plt
import pypsa
import cartopy.crs as ccrs
```

```python
n_nodes = 15
regions_method = 'a5'
```

```python
countries = gpd.read_file("resources/country_shapes.geojson")
europe = gpd.read_file("resources/europe_shape.geojson")
ren_zones = gpd.read_file(f"resources/renewable_shapes_{regions_method}.geojson")
```

```python
n = pypsa.Network(f"networks/elec_s_{n_nodes}_{regions_method}.nc")
```

```python
ren_z_col = matplotlib.collections.LineCollection(ren_zones.boundary)
```

```python
fig,axs = plt.subplots(1,subplot_kw={"projection":ccrs.PlateCarree()})
n.plot(ax = axs,geomap=True)
axs.add_collection(ren_z_col)
```

```python
fig.savefig(f"results/plots/Nodes_with_ren_zones_{n_nodes}_{regions_method}",dpi = 1000)
```
